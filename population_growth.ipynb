```python
#Importations
import matplotlib.pyplot as plt
```

```python
#Declare the variables
p_initial= 10000 #initial population
growth_rate= 0.07 #growth rate
time= 20 #in years

# Implementation using for loop
pop_lst= []
year_lst= []
for i in range(time):
  pop_lst.append(p_initial)
  year_lst.append(i)
  p_initial= p_initial*(1+growth_rate)

plt.plot(year_lst, pop_lst)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth Over Time')

```

```python
#Implementation with while loop

#Declare the variables
p_initial= 10000 #initial population
growth_rate= 0.07 #growth rate
time= 20 #in years

pop_lst_while= []
year_lst_while= []
year= 0
while time>0:
  pop_lst_while.append(p_initial)
  year_lst_while.append(year)
  p_initial= p_initial*(1+growth_rate)
  year+=1
  time-=1

plt.plot(year_lst_while, pop_lst_while)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth Over Time')
```
